#### Using more general functionality that uses any image and detect face on it.

```python
### Create a general functionality

def detect_face(img):
    face_img = img.copy()
    face_rec = face_cascade.detectMultiScale(face_img)
    for x,y,w,h in face_rec:
        cv2.rectangle(face_img,pt1=(x,y),pt2=(x+w,y+h),color=(255,255,255),thickness=10)
    
    return face_img
```

```python
plt.imshow(detect_face(nadia),cmap='gray')
```

```python
plt.imshow(detect_face(slovay),cmap='gray')
```

##### It is obvious that the algorithm works fine but still some wrong detection happens, also few double detection has been identified. So need to add few more secondary parameters.

Introducing scale factor and minimum neighbours -

    1. scaleFactor: This parameter compensates for the fact that an object at different distances from the camera will appear at different sizes. It controls how much the image size is reduced at each image scale. It must be strictly greater than 1. A lower scaleFactor increases the detection time but also increases the chance of detection. Typical values range from 1.01 to 1.3.

    2. minNeighbors: This parameter specifies how many neighbors each candidate object should have to retain it. Higher values result in fewer detections but with higher quality. Lower values may lead to more detections but with possible false positives. Its a trade-off between precision and recall.
    It means if multiple numbers of rectangles are near a certain area or a minimum number of neighours then we decide that is where the face is.

```python
### Modified general functionality

def adj_detect_face(img):
    face_img = img.copy()
    face_rec = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    for x,y,w,h in face_rec:
        cv2.rectangle(face_img,pt1=(x,y),pt2=(x+w,y+h),color=(255,255,255),thickness=10)
    
    return face_img
```

```python
adj_slovay = adj_detect_face(slovay)
```

```python
plt.imshow(adj_slovay)
```

### Detect the Eye

```python
eye_cascade = cv2.CascadeClassifier('DATA/haarcascades/haarcascade_eye.xml')
```

```python
### Create a general functionality

def detect_eyes(img):
    face_img = img.copy()
    eye_rec = eye_cascade.detectMultiScale(face_img)
    for x,y,w,h in eye_rec:
        cv2.rectangle(face_img,pt1=(x,y),pt2=(x+w,y+h),color=(255,255,255),thickness=10)
    
    return face_img
```
