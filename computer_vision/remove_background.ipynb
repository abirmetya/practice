```python
import cv2
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
```

```python
img2     = cv2.imread("D:/abir/udemy/computer_vision/eiffel_tower.jpg")
### CV2 reads img in the order og BGR and matplotlib expects image in the order og RGB
# CV2 ----> BGR
# matplotlib -----> RGB
fix_img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img2)
```

```python
## Mask an Image
# Step 1 --> Convert to gray scale
img2gray = cv2.cvtColor(fix_img2,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
# Step 2 --> Take inverse of the image

mask_inv = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv,cmap='gray')
```

```python
# threshold on white
# Define lower and uppper limits
lower = np.array([200])#, 200, 200])
upper = np.array([255])#, 255, 255])

# Create mask to only select black
thresh = cv2.inRange(mask_inv, lower, upper)
```

```python
white_background = np.full(fix_img2.shape,255,dtype=np.uint8)

bk = cv2.bitwise_or(white_background,white_background,mask=thresh)
```

```python
plt.imshow(bk)
```
